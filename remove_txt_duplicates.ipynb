```python
# 删除txt中的重复行

import os
import tqdm
def read_txt_file(file_path: str) -> str:
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    return content


```

```python
import difflib

def detect_duplicates(content: str):
    lines = content.splitlines()
    duplicates = set()
    
    for i in tqdm.tqdm(range(len(lines))):
        for j in range(i + 1, len(lines)):
            if 10 < len(lines[i]) < 100 and 10 < len(lines[j]) < 100:
                edit_distance = difflib.SequenceMatcher(None, lines[i], lines[j]).ratio()
                if (1 - edit_distance) * len(lines[i]) < 1:  # 计算编辑距离
                    duplicates.add(lines[j])
    
    return duplicates

import re

def break_lines(lines: list, split_word):
    new_lines = []
    for line in lines:
        # 使用正则表达式分割行，并保留分割词
        new_lines.extend(re.split(f'({re.escape(split_word)})', line))
    return new_lines


# detect totally indentical duplicates
def detect_totally_identical_duplicates(content: str):
    lines = content.splitlines()
    # break lines with 'com' or 'COM' and make a new list
    lines = break_lines(lines, 'com')
    lines = break_lines(lines, 'COM')
    lines = break_lines(lines, 'org')
    lines = break_lines(lines, 'ORG')

    duplicates = set()
    for i in tqdm.tqdm(range(len(lines))):
        for j in range(i + 1, len(lines)):
            if lines[i] == lines[j]:
                duplicates.add(lines[j])
    return duplicates

```

```python
def detect_duplicate_substrings(content: str, min_length: int = 15, max_length: int = 20) -> set:
    seen = set()
    duplicates = set()
    
    for length in tqdm.tqdm(range(min_length, max_length)):  # 子字符串长度从min_length到一半的内容长度
        for start in range(len(content) - length + 1):
            substring = content[start:start + length]
            if substring in seen:
                duplicates.add(substring)
            else:
                seen.add(substring)
    
    return duplicates
```

```python
file_path = r"C:\projects\data\qingmingxueyuan.txt"
lines = read_txt_file(file_path)
#duplicates = detect_totally_identical_duplicates(lines)
# add contents to a single string
content = ''.join(lines.splitlines())
duplicates = detect_duplicate_substrings(content, min_length=15, max_length=20)
print(len(duplicates))
    
```

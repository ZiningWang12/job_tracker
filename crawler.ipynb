```python
import requests
from bs4 import BeautifulSoup
import re

base_url = 'https://www.cool18.com/bbs4/index.php?app=forum&act=threadview&tid=14039813'
response = requests.get(base_url)

def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    return True

def get_valid_text(soup):
    texts = soup.find_all(text=True)
    visible_texts = filter(tag_visible, texts)
    all_text = [t.strip() for t in visible_texts]
    return ' '.join(all_text)

def crawl_links(soup):
    links = soup.find_all('a', href=True)
    valid_texts = []
    for link in links:
        if '清茗学院' in link.text:
            link_url = link['href']
            link_response = requests.get(link_url)
            if link_response.status_code == 200:
                link_soup = BeautifulSoup(link_response.text, 'html.parser')
                valid_text = get_valid_text(link_soup)
                valid_texts.append(valid_text)
    return valid_texts

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    texts = crawl_links(soup)
    longest_text = max(texts, key=len) if texts else ''
    print(longest_text)
else:
    print(f"请求失败，状态码: {response.status_code}")


```

```python
def tag_visible(element):
    # 过滤掉不需要的标签
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    return True

def get_valid_text(response):
    # 提取所有文本内容
    texts = response.find_all(text=True)
    # 过滤掉不需要的标签
    visible_texts = filter(tag_visible, texts)
    # 合并所有可见文本
    all_text = [t for t in visible_texts]
    valid_lines = []
    for i, t in enumerate(all_text):
        if 'cool18.com' in t and i > 0:
            valid_lines.append(all_text[i-1].strip('\r\n'))
            last_line = all_text[i+1].strip()
    valid_lines.append(last_line)
    valid_text = ''.join(valid_lines)
    return valid_text

def crawl_links(soup):
    links = soup.find_all('a', href=True)
    valid_texts = {}
    print(len(links))
    for link in tqdm.tqdm(links[10:199]):
        if '清茗学院' in link.text:
            link_url = link['href']
            link_response = requests.get(link_url)
            if link_response.status_code == 200:
                link_soup = BeautifulSoup(link_response.text, 'html.parser')
                valid_text = get_valid_text(link_soup)
                valid_texts[link.text] = valid_text
    return valid_texts

def crawl_own_text(soup):
    # 提取所有文本内容
    texts = soup.find_all(text=True)
    # 过滤掉不需要的标签
    visible_texts = filter(tag_visible, texts)
    # 合并所有可见文本
    all_text = [t for t in visible_texts]
    valid_text = get_valid_text(soup)
    return valid_text, texts, all_text
```

```python
import requests
from bs4 import BeautifulSoup
import re
import tqdm

url = 'https://www.cool18.com/bbs4/index.php?app=forum&act=threadview&tid=14039813'
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    texts = crawl_links(soup)
    texts['第一章'] = get_valid_text(soup)
else:
    print(f"请求失败，状态码: {response.status_code}")
# save texts to dictionary file
import json
with open('qingmingxueyuan1.json', 'w') as f:
    json.dump(texts, f)
```
